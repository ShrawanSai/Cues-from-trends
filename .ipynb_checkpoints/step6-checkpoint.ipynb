```python
from nltk.tokenize import word_tokenize
```

```python
text = " SidNaaz Craze and Burnol Trending together üòÇü§£"
print(word_tokenize(text))
```

```python
import gensim
all_stopwords = set(gensim.parsing.preprocessing.STOPWORDS)
#print(set(all_stopwords))
```

```python
line = " #SidNaaz Craze and Burnol Trending together üòÇü§£ !!"
```

```python
#gather_details
line = word_tokenize(line)
exclamation_indices = []
emoji_indices = []
hashtag_indices = []
named_entities = {}

c = 0
for word_index in range(len(line)):
    if line[word_index] in all_stopwords:
        continue
    elif line[word_index].startswith('#'):
        hashtag_indices.append(word_index)
        
    elif line[word_index] == '!':
        exclamation_indices.append(word_index)
    
    elif is_emoji(line[word_index][0]) :
        emoji_indices.append(word_index)
      
print(exclamation_indices,emoji_indices, hashtag_indices)
```
